```python
#import dependencies
import pandas as pd
import matplotlib as plt
```

```python
# read csv file downloaded from open data sfgov site
csv_path = "Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv"
data = pd.read_csv(csv_path)

#explore data 
data.head(100)
```

```python
#display column headers
data.columns
```

```python
# create a reduced dataframe with only the columns relevant
df = data[['Category', 'Descript', 'DayOfWeek', 'Date', 'Time', 'PdDistrict', 'X', 'Y', 'PdId']]
df.head()
```

```python
# Parse the 'Date' column for year and create a new column for Year
df['Date'] = pd.to_datetime(df.Date)
df['Year'] = df.Date.dt.year

df.head()
```

```python
#Filter the dataframe for Larceny/Theft incidents only
df_ld = df.loc[df["Category"] == "LARCENY/THEFT"]
df_ld

#Save Larceny/Theft data to a new csv
df_ld.to_csv('data larceny and theft with year label.csv')
```

```python
# Visualization # 1. Line chart to show trend of larceny/theft over time in the entire dataset.
import matplotlib.pyplot as plt
df_ld.Year.value_counts().sort_index().plot(figsize =(15,15))
plt.show()
```

```python
import pandas as pd
df_sf=pd.read_csv('df_ld.csv') 
df_cg=pd.read_csv('chicago.csv') 
```

```python
df_sf = df_sf.loc[(df_sf['Year'] <= 2017) & (df_sf['Year'] >= 2003)]
df_cg = df_cg.loc[(df_cg['Year'] <= 2017) & (df_cg['Year'] >= 2003)]
```

```python
import matplotlib.pyplot as plt
data_sf = df_sf.Year.value_counts().sort_index()
data_cg = df_cg.Year.value_counts().sort_index()
plt.show()
```

```python
plt.plot(data_sf, label = "San Francisco")
plt.plot(data_cg, label = "Chicago")
plt.title('Total number of Larceny/Theft from 2003 to 2017 in San Francisco Vs. Chicago')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid()
plt.legend()
plt.savefig('San Francisco Vs. Chicago.png')
```

```python
df=pd.read_csv('df_ld.csv', low_memory=False)  # read geocoded location from file
df.columns
```

```python
df = df[['X', 'Y']]
sample = df.sample(50000)
sample.head()     
```

```python
sample['weight'] = 1
sample.head()
```

```python
locations = sample[['Y', 'X']]          # put latitide and longitude as a variable name 'locations'
val = sample['weight']                                 # put the weight into variable name 'val'
locations.head()                                          # display the first 5 rows of the location data
```

```python
def drawHeatMap(location, val, zoom, intensity, radius):
    # setting the data and parameters
    heatmap_layer = gmaps.heatmap_layer(locations, val, dissipating = True)
    heatmap_layer.max_intensity = intensity
    heatmap_layer.point_radius = radius
    # draw the heatmap into a figure
    fig = gmaps.figure()
    fig = gmaps.figure(center = [37.75,-122.4], zoom_level=zoom)
    fig.add_layer(heatmap_layer)
    return fig
```

```python
# set up parameters
zoom=12
intensity=1
radius=2

# call the function to draw the heatmap
drawHeatMap(locations, val, zoom, intensity, radius)
```
